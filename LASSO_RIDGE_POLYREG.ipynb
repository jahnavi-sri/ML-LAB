```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('horror_movies.csv')
data=data.dropna()
data
```

```python

X = data[['popularity'] ]
y = data['budget']
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6)
```

```python

lasso_model = Lasso(alpha=1.0)
lasso_model.fit(X_train, y_train)
lasso_predictions = lasso_model.predict(X_test)


```

```python
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
ridge_predictions = ridge_model.predict(X_test)
```

```python
degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

poly_regression = Ridge(alpha=1.0)
poly_regression.fit(X_train_poly, y_train)
poly_predictions = poly_regression.predict(X_test_poly)
```

```python
def calculate_metrics(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return r2, mae, rmse


```

```python
lasso_r2, lasso_mae, lasso_rmse = calculate_metrics(y_test, lasso_predictions)
ridge_r2, ridge_mae, ridge_rmse = calculate_metrics(y_test, ridge_predictions)
poly_r2, poly_mae, poly_rmse = calculate_metrics(y_test, poly_predictions)

```

```python

print("Lasso Regression - R2:", lasso_r2, "MAE:", lasso_mae, "RMSE:", lasso_rmse)
print("Ridge Regression - R2:", ridge_r2, "MAE:", ridge_mae, "RMSE:", ridge_rmse)
print("Polynomial Regression - R2:", poly_r2, "MAE:", poly_mae, "RMSE:", poly_rmse)
```
